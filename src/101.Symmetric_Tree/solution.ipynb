解题思路: 

1. 首先我们看题目, 是要求判断根节点的左右两棵子树是否镜像对称.

2. 若 root 为空, 那么返回 True.

3. 我们将根节点的左子树记做 left, 右子树记做 right. 那么我们现在就有了三个终止条件了:

    - left 和 right 都为空 ==> 返回 True

    - left 和 right 其中有一个为空 ==> 返回 False 

    - left 和 right 的 val 不等 ==> 返回 False 
    
3. 如果相当, 那么我们递归比较 left 的左节点和 right 的右节点, 再比较 left 的右节点和 right 的左节点, 看看是否相等.


`````python
# Definition for a binary tree node.
# class TreeNode(object):
#     def __init__(self, x):
#         self.val = x
#         self.left = None
#         self.right = None

class Solution(object):
    def isSymmetric(self, root):
        """
        :type root: TreeNode
        :rtype: bool
        """
        if not root:
            return True
        
        def dfs(left,right):
            # 如果左右节点都为空, return True
            # 如果左右节点只有一个为空, return False
            # 若左节点的 val 不等于右节点的 val, return False
            if not (left or right):
                return True
            elif not (left and right):
                return False
            elif left.val != right.val:
                return False
            
            return dfs(left.left,right.right) and dfs(left.right,right.left)
        
        return dfs(root.left,root.right)
`````

除了使用递归的方法, 我们还以使用队列来解决这个问题.

1. 我们还是首先判断根节点是否为空, 若为空的返回 True.

2. 我们使用一个列表来模拟队列, 我们初始化放入根节点的左子树和右子树.

3. 当队列非空, 我们进入 while 循环.
    - 我们首先从队列中提取左右节点, 再比较这两个节点

    - 若两个节点均为空就继续循环, 若两者中有一个为空或者值不想等就返回 False.

    - 将左节点的左child和右节点的右child放入队列

    - 将左节点的右child和右节点的左child放入队列

4. 若循环结束也没有返回任何值, 那么返回 True

`````python
# Definition for a binary tree node.
# class TreeNode(object):
#     def __init__(self, x):
#         self.val = x
#         self.left = None
#         self.right = None

class Solution(object):
    def isSymmetric(self, root):
        """
        :type root: TreeNode
        :rtype: bool
        """
        if not root:
            return True
        
        queue = [root.left,root.right]

        while queue:
            left = queue.pop(0)
            right = queue.pop(0)

            if not (left or right):
                continue
            elif not (left and right):
                return False
            elif left.val != right.val:
                return False
            
            queue.append(left.left)
            queue.append(right.right)

            queue.append(left.right)
            queue.append(right.left)
        
        return True
`````

Note: 如果题目需要先遍历到树最底层再处理数据的话用栈，如果需要边遍历边处理数据就用队列。
